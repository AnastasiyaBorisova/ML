```python
import copy


data = [['–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞', '–û—Å–∞–¥–∫–∏', '–ú–µ—Å—è—Ü', '–ò–≥—Ä–∞'],
        ['< 0', '–î–∞', '–ê–ø—Ä–µ–ª—å', '–ù–µ—Ç'],
        ['< 0', '–î–∞', '–ú–∞—Ä—Ç', '–ù–µ—Ç'],
        ['10 ‚àí 15', '–ù–µ—Ç', '–ú–∞–π', '–î–∞'],
        ['> 15', '–ù–µ—Ç', '–ú–∞–π', '–î–∞'],
        ['< 0', '–ù–µ—Ç', '–ê–ø—Ä–µ–ª—å', '–î–∞'],
        ['> 15', '–î–∞', '–ú–∞—Ä—Ç', '–ù–µ—Ç'],
        ['10 ‚àí 15', '–ù–µ—Ç', '–ú–∞–π', '–î–∞'],
        ['< 10', '–î–∞', '–ú–∞–π', '–î–∞'],
        ['< 10', '–ù–µ—Ç', '–ê–ø—Ä–µ–ª—å', '–î–∞'],
        ['> 15', '–ù–µ—Ç', '–ú–∞–π', '–î–∞'],
        ['< 10', '–î–∞', '–ú–∞—Ä—Ç', '–ù–µ—Ç'],
        ['10 ‚àí 15', '–ù–µ—Ç', '–ú–∞–π', '–î–∞'],
        ['10 ‚àí 15', '–ù–µ—Ç', '–ê–ø—Ä–µ–ª—å', '–ù–µ—Ç'],
        ['> 15', '–ù–µ—Ç', '–ú–∞–π', '–ù–µ—Ç']]
valuex = []
tableP = []
table = []
pc = []
xck = []
ver = []

def choice(data, question): #–í—ã–±–æ—Ä –∫—Ä–∏—Ç–µ—Ä–∏—è
    for i in range(len(data)):
        print(i, ':', data[i])
    while (True):
        num = int(input(question))
        if num >= 0 and num < len(data):
            select = data[int(num)]
            break
        else:
            print("–û—à–∏–±–∫–∞!")
            exit
    return select

def multipy(m):
    p = 1
    for i in m:
        p *= i
    return p

pr = choice(data[0], '–ö–∞–∫–æ–π –∫—Ä–∏—Ç–µ—Ä–∏–π —Å–ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º? ')
other = copy.deepcopy(data[0])
other.remove(pr)

pro = sorted(set(list(zip(*data[1:]))[data[0].index(pr)])) #–í–∞—Ä–∏–∞–Ω—Ç—ã –∑–Ω–∞—á–µ–Ω–∏–π –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º–æ–≥–æ –∫—Ä–∏—Ç–µ—Ä–∏—è
for el in other: #–í—ã–±–æ—Ä –∑–Ω–∞—á–µ–Ω–∏–π –≤ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –∫—Ä–∏—Ç–µ—Ä–∏—è—Ö –¥–ª—è X
    valuex.append(choice(sorted(set(list(zip(*data[1:]))[data[0].index(el)])), '–í—ã–±–µ—Ä–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫—Ä–∏—Ç–µ—Ä–∏—è ' + el +': '))

print('\nX =', valuex)

for i in range(len(pro)): #–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å ùëÉ(ùê∂ùëò) (–°–∫–æ–ª—å–∫–æ —Ä–∞–∑ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –∫–∞–∂–¥—ã–π –∏–∑ –∫–ª–∞—Å—Å–æ–≤)
    tableP.append(round(list(zip(*data[1:]))[data[0].index(pr)].count(pro[i]) / (len(data) - 1), 5))
    print('P(', pr, '=', pro[i], ') = ', tableP[i])

for i in range(len(other)): #–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å (–°–∫–æ–ª—å–∫–æ —Ä–∞–∑ –≤ –≤—ã–±—Ä–∞–Ω–Ω–æ–º –∫—Ä–∏—Ç–µ—Ä–∏–∏ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –∫–∞–∂–¥–æ–µ –∏–∑ –∑–Ω–∞—á–µ–Ω–∏–π –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º–æ–≥–æ –∫–ª–∞—Å—Å–∞)
    table.append([])
    for j in range(len(pro)):
        a = 0
        for k in range(1, len(data)):
            if data[k][data[0].index(pr)] == pro[j] and data[k][data[0].index(other[i])] == valuex[i]:
                a += 1
        b = round(a / list(zip(*data[1:]))[data[0].index(pr)].count(pro[j]), 5)
        print('P(', other[i],'=', valuex[i], '|', pr, '=', pro[j], ') = ', b)
        table[i].append(b)
table = list(zip(*table))

for i in range(len(pro)): #–ü–µ—Ä–µ–º–Ω–æ–∂–∞–µ–º ùëÉ(ùëã|ùê∂k)
    pc.append(round(multipy(table[i]), 5))
    print('P( X |', pr, '=', pro[i], ') =', pc[i])

for i in range(len(pc)): #ùëÉ(ùëã|ùê∂k) √ó ùëÉ(ùê∂k)
    xck.append(round(pc[i] * tableP[i], 5))
    print('P( X |', pr, '=', pro[i], ') * P(', pr, '=', pro[i], ') =', xck[i])

for i in range(len(xck)): #–ù–æ—Ä–º–∞–ª–∏–∑—É–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
    ver.append(round(xck[i] / sum(xck), 5))
    print('–í–µ—Ä–æ—è–Ω–æ—Å—Ç—å –∏—Å—Ö–æ–¥–∞ (', pr, '=', pro[i],') = ', ver[i])

print('–î–ª—è –¥–∞–Ω–Ω–æ–≥–æ X –Ω–∞–∏–±–æ–ª–µ–µ –≤–µ—Ä–æ—è—Ç–Ω—ã–π –∏—Å—Ö–æ–¥', pr, '=', pro[xck.index(max(xck))], '(', round(max(ver) * 100, 1), '% )')
```
